```python
import numpy as np
import matplotlib.pyplot as plt

# Create data
x = ['Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul',
     'Aug', 'Sep', 'Oct', 'Nov', 'Dec', 'Jan']
y = [54, 47, 81, 84, 85, 85, 82, 86, 84, 85, 86, 80]


# Area plot

plt.fill_between(x, y, color="#003f5c")
plt.plot(x, y, color="#ff6361")

plt.title(
    "Share of people who wore face masks during COVID-19 in Indonesia", loc="center")
plt.xlabel("Month (2020-21)")
plt.ylabel("Percentage")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# set width of bars
barWidth = 0.25


bars1 = [380, 220, 16]
bars2 = [220, 14, 10]

x = np.arange(len(bars1))

r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]


labels = ['Medical Face Mask', 'Antiviral Drugs', 'Rotivoviral Drugs']
bars1 = [380, 220, 16]
bars2 = [220, 14, 10]

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, bars1, width,
                label='Sales Volume Growth', color='#58508d')
rects2 = ax.bar(x + width/2, bars2, width,
                label='Average Bill Growth', color='#ff6361')

ax.set_ylabel('Percentage')
ax.set_title('Growth of COVID Preventive Products from Feb 2020 to March 2020')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


fig.tight_layout()

plt.show()
```
